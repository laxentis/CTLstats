```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_json('data.json', orient='year')
EUDcandidate = {}
EUDparticipant = {}
MENAcandidate = {}
MENAparticipant = {}
fir_EUDcandidate = {}
fir_EUDparticipant = {}
fir_MENAcandidate = {}
fir_MENAparticipant = {}
```

```python
print(data['year'])
```

```python
def icao_to_fir(icao):
    return icao[0:2]
```

```python
for row in data.candidates:
    for airport in row['arabia']:
        if airport not in MENAparticipant:
            MENAparticipant[airport] = 0
            fir_MENAparticipant[icao_to_fir(airport)] = 0
        if airport not in MENAcandidate:
            MENAcandidate[airport] = 1
            fir_MENAcandidate[icao_to_fir(airport)] = 1
        else:
            MENAcandidate[airport] += 1
            fir_MENAcandidate[icao_to_fir(airport)] += 1
    for airport in row['europe']:
        if airport not in EUDparticipant:
            EUDparticipant[airport] = 0
            fir_EUDparticipant[icao_to_fir(airport)] = 0
        if airport not in EUDcandidate:
            EUDcandidate[airport] = 1
            fir_EUDcandidate[icao_to_fir(airport)] = 1
        else:
            EUDcandidate[airport] += 1
            fir_EUDcandidate[icao_to_fir(airport)] += 1
```

```python
for row in data.chosen:
    for airport in row['arabia']:
        MENAparticipant[airport] += 1
        fir_MENAparticipant[icao_to_fir(airport)] += 1
    for airport in row['europe']:
        EUDparticipant[airport] += 1
        fir_EUDparticipant[icao_to_fir(airport)] += 1
```

```python
EUDparticipant = dict(sorted(EUDparticipant.items()))
EUDcandidate = dict(sorted(EUDcandidate.items()))
MENAparticipant = dict(sorted(MENAparticipant.items()))
MENAcandidate = dict(sorted(MENAcandidate.items()))

fir_EUDparticipant = dict(sorted(fir_EUDparticipant.items()))
fir_EUDcandidate = dict(sorted(fir_EUDcandidate.items()))
fir_MENAparticipant = dict(sorted(fir_MENAparticipant.items()))
fir_MENAcandidate = dict(sorted(fir_MENAcandidate.items()))
```

```python
plt.figure(figsize=(20, 6))
plt.bar(range(len(EUDcandidate)), list(EUDcandidate.values()), align='center', label='Candidate')
plt.bar(range(len(EUDparticipant)), list(EUDparticipant.values()), align='center', label='Chosen')
plt.xticks(range(len(EUDparticipant)), list(EUDparticipant.keys()), rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
plt.bar(range(len(fir_EUDcandidate)), list(fir_EUDcandidate.values()), align='center', label='Chosen')
plt.bar(range(len(fir_EUDparticipant)), list(fir_EUDparticipant.values()), align='center', label='Candidate')
plt.xticks(range(len(fir_EUDcandidate)), list(fir_EUDcandidate.keys()), rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
plt.bar(range(len(MENAcandidate)), list(MENAcandidate.values()), align='center', label='Candidate')
plt.bar(range(len(MENAparticipant)), list(MENAparticipant.values()), align='center', label='Chosen')
plt.xticks(range(len(MENAparticipant)), list(MENAparticipant.keys()), rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 6))
plt.bar(range(len(fir_MENAcandidate)), list(fir_MENAcandidate.values()), align='center', label='Candidate')
plt.bar(range(len(fir_MENAparticipant)), list(fir_MENAparticipant.values()), align='center', label='Chosen')
plt.xticks(range(len(fir_MENAcandidate)), list(fir_MENAcandidate.keys()), rotation=90)
plt.legend()
plt.show()
```
